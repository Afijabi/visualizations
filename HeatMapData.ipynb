```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("hgmd.raw.matrix.csv")
```

```python
labels = [None] * len(data)
for i in range(0, len(data)):
    labels[i] = str(data['phenotype'][i]) + " (chr" + str(data['CHROM_hg19'][i]) + ":" + str(data['POS_hg19'][i]) + ")"
```

```python
data1 = data.iloc[:,9:]
data1.drop(data1.iloc[:, 1:3], axis = 1, inplace = True)
data1['phenotype'] = labels
data2 = data1.set_index('phenotype')
```

```python
fig, ax = plt.subplots(figsize=(30, 300))
ax = sb.heatmap(data2, cmap = "RdBu_r", vmin = -8, vmax = 8, square=True,
          linewidth=0.01, cbar_kws={"shrink": .01})

ax.xaxis.tick_top()

plt.xticks(rotation=90)
plt.show()
```
