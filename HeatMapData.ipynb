```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("hgmd.raw.matrix.csv", usecols = [i for i in range(9, 52) if i not in range(10, 12)], index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(30, 300))
ax = sb.heatmap(data, cmap = "RdBu_r", vmin = -11, vmax = 11, square=True,
          linewidth=0.01, cbar_kws={"shrink": .01})

ax.xaxis.tick_top()

plt.xticks(rotation=90)
plt.show()
```
